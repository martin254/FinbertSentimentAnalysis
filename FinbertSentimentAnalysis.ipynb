```python
# Step 2: Import necessary modules
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from tqdm import tqdm  # For progress tracking

# Load your dataset

print(f"Original dataset shape: {stocks.shape}")
```

```python
# Step 3: Initialize FinBERT model and tokenizer
model_name = "ProsusAI/finbert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Define label mapping (from model documentation)
label_map = {
    0: "positive",
    1: "negative",
    2: "neutral"
}
```

```python
# Step 4: Create sentiment analysis function with batch processing
def analyze_sentiment_batch(texts, batch_size=8):
    results = []
    
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        
        # Tokenize batch
        inputs = tokenizer(
            batch.tolist(),
            padding=True,
            truncation=True,
            max_length=128,  # Optimal for headlines
            return_tensors="pt"
        )
        
        # Model prediction
        with torch.no_grad():
            outputs = model(**inputs)
        
        # Get probabilities
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        
        # Store results
        for j in range(len(batch)):
            results.append({
                'sentiment': label_map[probs.argmax(dim=-1)[j].item()],
                'positive': probs[j][0].item(),
                'negative': probs[j][1].item(),
                'neutral': probs[j][2].item()
            })
    
    return results
```

```python
# Step 5: Process the titles in batches
titles = stocks['Title'].fillna('')  # Handle missing values
sentiment_results = analyze_sentiment_batch(titles)

# Convert results to DataFrame
sentiment_df = pd.DataFrame(sentiment_results)

# Combine with original data
final_df = pd.concat([stocks, sentiment_df], axis=1)
```
